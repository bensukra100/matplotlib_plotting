```python
## modules
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import MultipleLocator
import os
markers = mpl.markers.MarkerStyle.filled_markers


##output directory
out_dir = 'sample_plots'
os.makedirs('C:\\Users\\bsukra\\repositories\\personal\\matplotlib_plotting\\'+out_dir)
output_directory = 'C:\\Users\\bsukra\\repositories\\personal\\matplotlib_plotting\\'+out_dir

## matplotlib text/marker features
mpl.rc('font',family='serif')
fsize = 23
lnwidth = 1
tickwidth = 1
msize = 7
```

```python
## fake data
x = np.linspace(-4.*np.pi, 4.*np.pi, 1000)
y0 = np.sin(x)
y1 = np.cos(x)
```

```python
## PLOTTING

## figure set-up
fig = plt.figure(figsize=(12,10))
spec = mpl.gridspec.GridSpec(ncols=1, nrows=1)
# spec = mpl.gridspec.GridSpec(ncols=2, nrows=2, height_ratios=(4,1), width_ratios=(4,1), wspace=0.0, hspace=0.0)

## add a subplot
ax0 = fig.add_subplot(spec[0, 0])

## plotting the data 
ax0.plot(x, y0, color='blue', lw=3, alpha=0.8, linestyle='-.', label='sine')
ax0.plot(x, y1, color='red', lw=3, alpha=0.8, linestyle='-', label='cosine')

## AESTHETICS
## labelling 
ax0.set_xlabel('x label', fontsize=fsize)
ax0.set_ylabel('y label', fontsize=fsize)
ax0.legend(loc=(0.75,0), fontsize = fsize, frameon=0)

## spines
################################
ax0.spines['right'].set_linewidth(lnwidth*2)
ax0.spines['top'].set_linewidth(lnwidth*2)
ax0.spines['left'].set_linewidth(lnwidth*2)
ax0.spines['bottom'].set_linewidth(lnwidth*2)
################################

## tick parameters
################################
ax0.tick_params(axis='both', which='major', labelsize=fsize*0.9, width = tickwidth*2, length = 11, direction='in',
               top='True', bottom='True',left='True',right='True')
ax0.tick_params(axis='both', which='minor', width = tickwidth*2, length = 5.5, direction='in',
               top='True', bottom='True',left='True',right='True')
#################################

## ticks
################################
m0 = MultipleLocator(0.25)
ax0.yaxis.set_major_locator(m0)
m0 = MultipleLocator(0.05)
ax0.yaxis.set_minor_locator(m0)
################################

################################
m0 = MultipleLocator(5)
ax0.xaxis.set_major_locator(m0)
m0 = MultipleLocator(1)
ax0.xaxis.set_minor_locator(m0)
################################

## dropping some lines 
## horizontal
ax0.axhline(-np.sqrt(2.)/2., c='k', lw=3, alpha = 0.4)
ax0.axhline(np.sqrt(2.)/2, c='k', lw=3, alpha = 0.4) 
ax0.axhline(0, c='k', lw=3, alpha=0.5, linestyle='-.')
## vertical 
ax0.axvline(-3.*np.pi/4., c='k', lw=3, alpha = 0.4)
ax0.axvline(np.pi/4., c='k', lw=3, alpha = 0.4) 

## axis limits
ax0.set_xlim(-10, 10)
ax0.set_ylim(-1.25, 1.25)

## SAVING IMAGE
fig.tight_layout()
#############################################################################################
fname = "sampleplot_0.pdf"
fileout = os.path.join(output_directory, fname)

plt.savefig(fileout,dpi=300, bbox_inches="tight");
```
